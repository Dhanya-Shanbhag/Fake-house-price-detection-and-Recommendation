```python

# Perform one-hot encoding on the 'STATUS' column with prefix
df2_encoded_status = pd.get_dummies(df2, columns=['STATUS'], prefix='STATUS', drop_first=True)

# Rename the new columns to remove the prefix
df2_encoded_status.rename(columns={'STATUS_Under Construction': 'STATUS'}, inplace=True)

# Display the modified DataFrame
display(df2_encoded_status)
```

# BUILD A RECOMMENDATION SYSTEM

```python
df_comb = pd.merge(df_standardized, df2_encoded[['ID']], left_index=True, right_index=True)

# Display the resulting dataframe
display(df_comb)
```

```python
# Merge the two dataframes based on their indices
df_combined = pd.merge(df_rec, df2_encoded[['ID']], left_index=True, right_index=True)

# Display the resulting dataframe
display(df_combined)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets (adjust the test_size as needed)
df_train, df_test = train_test_split(df_combined, test_size=0.3, random_state=42)
```

```python
df_test
```

```python
from sklearn.neighbors import NearestNeighbors

# Extract features from df_train and df_test
X_train = df_train.drop(['ID'], axis=1)
X_test = df_test.drop(['ID'], axis=1)

# Define the number of neighbors (in this case, 10)
n_neighbors = 10

# Initialize an empty DataFrame to store neighbor details
neighbor_columns = [f'Neighbor_{i}' for i in range(1, n_neighbors + 1)]
df_test_with_neighbors = pd.concat([df_test, pd.DataFrame(columns=neighbor_columns)], axis=1)

# Fit the KNN model on the training data
knn_model = NearestNeighbors(n_neighbors=n_neighbors)
knn_model.fit(X_train)

# Loop through each record in df_test
for i in range(len(df_test)):
    # Find the 10 nearest neighbors for the current record
    distances, indices = knn_model.kneighbors(X_test.iloc[[i]])
    
    # Extract the IDs of the neighbors
    neighbor_ids = df_train.iloc[indices[0]]['ID'].tolist()
    
    # Assign the neighbor IDs to the corresponding columns in df_test_with_neighbors
    df_test_with_neighbors.loc[df_test.index[i], neighbor_columns] = neighbor_ids

# Display the updated dataframe
display(df_test_with_neighbors)
```

```python
desired_ids = [2723, 2679]

# Extract records with the specified IDs
selected_records = df2_encoded.loc[df2_encoded['ID'].isin(desired_ids)]

# Display the selected records
display(selected_records)
```

```python

# Initialize empty lists to store Min, Max, and Average values
min_prices = []
max_prices = []
avg_prices = []

# Iterate through each row in df_test_with_neighbors
for i in range(len(df_test_with_neighbors)):
    # Extract the neighbor IDs from the current row
    neighbor_ids = df_test_with_neighbors.iloc[i, -n_neighbors:].tolist()
    
    # Look up the PRICE values in df2_encoded for the neighbor IDs
    neighbor_prices = df2_encoded.loc[df2_encoded['ID'].isin(neighbor_ids)]['PRICE'].tolist()
    
    # Calculate Min, Max, and Average
    min_price = min(neighbor_prices) if neighbor_prices else None
    max_price = max(neighbor_prices) if neighbor_prices else None
    avg_price = sum(neighbor_prices) / len(neighbor_prices) if neighbor_prices else None
    
    # Append the values to the respective lists
    min_prices.append(min_price)
    max_prices.append(max_price)
    avg_prices.append(avg_price)

# Create a new DataFrame Calculated_df
calculated_df = df_test_with_neighbors.copy()

# Add the Min, Max, and Average columns to Calculated_df
calculated_df['Min'] = min_prices
calculated_df['Max'] = max_prices
calculated_df['Average'] = avg_prices

# Display the updated dataframe
display(calculated_df)
```

```python
import pandas as pd

def house_recommendation_system(user_id, num_recommendations, calculated_df, df2_encoded):
   
    recommendations = pd.DataFrame()

    # Validate the number of recommendations
    if 1 <= num_recommendations <= 10:
        # Find the record for the given ID in calculated_df
        user_records = calculated_df[calculated_df['ID'] == user_id]

        if not user_records.empty:
            # Take the first record if there are multiple matches
            user_record = user_records.iloc[0]

            # Retrieve the 'PRICE' value from df2_encoded for the given ID
            user_price = df2_encoded[df2_encoded['ID'] == user_id]['PRICE'].values[0]
            print()
            print('The price of input house is: ', user_price, 'L')
            print()
            print('The details of recommended houses are as below:')
            print()

            # Determine whether to recommend neighbors based on 'Min' or 'Average'
            if user_price < user_record['Average']:
                # Recommend neighbors with 'PRICE' values closer to 'Min'
                recommended_neighbors = user_record[['Neighbor_1', 'Neighbor_2', 'Neighbor_3', 'Neighbor_4', 'Neighbor_5',
                                                     'Neighbor_6', 'Neighbor_7', 'Neighbor_8', 'Neighbor_9', 'Neighbor_10']].tolist()

                # Retrieve details of recommended neighbors from df2_encoded
                recommended_neighbors_df = df2_encoded[df2_encoded['ID'].isin(recommended_neighbors)]

                # Find the differences between 'PRICE' and 'Min'
                recommended_neighbors_df['Difference'] = abs(recommended_neighbors_df['PRICE'] - user_record['Min'])

                # Sort by differences and keep the top recommendations
                recommendations = recommended_neighbors_df.sort_values(by='Difference').head(num_recommendations)

            else:
                # Recommend neighbors with 'PRICE' values around 'Average'
                recommended_neighbors = user_record[['Neighbor_1', 'Neighbor_2', 'Neighbor_3', 'Neighbor_4', 'Neighbor_5',
                                                     'Neighbor_6', 'Neighbor_7', 'Neighbor_8', 'Neighbor_9', 'Neighbor_10']].tolist()

                # Retrieve details of recommended neighbors from df2_encoded
                recommended_neighbors_df = df2_encoded[df2_encoded['ID'].isin(recommended_neighbors)]

                # Find the differences between 'PRICE' and 'Average'
                recommended_neighbors_df['Difference'] = abs(recommended_neighbors_df['PRICE'] - user_record['Average'])

                # Sort by differences and keep the top recommendations
                recommendations = recommended_neighbors_df.sort_values(by='Difference').head(num_recommendations)

        else:
            print(f"No records found for user ID {user_id} in calculated_df.")

    else:
        print("Number of recommendations must be between 1 and 10.")

    return recommendations

# Call the function
user_id_input = int(input("Enter the ID: "))
num_recommendations_input = int(input("Enter the number of recommendations (1-10): "))

result = house_recommendation_system(user_id_input, num_recommendations_input, calculated_df, df2_encoded)
display(result)
```

# IDENTIFICATION OF INACCURATELY LABELED HOUSE - PRICE

```python
# Creating a new dataframe 'price_label' to store the final output
price_label = pd.DataFrame(columns=['ID', 'Average', 'PRICE', 'Price label'])

for index, row in calculated_df.iterrows():
    # Search for the 'PRICE' value in df2_encoded based on 'ID'
    price_value = df2_encoded[df2_encoded['ID'] == row['ID']]['PRICE'].values

    # Check if 'PRICE' value is found
    if len(price_value) > 0:
        price_value = price_value[0]

        # Update the 'Price label' based on the condition
        if price_value >= row['Average'] + 6:
            label = 'overpriced' 
        else:
            label = 'correctly priced'

        # Append the result to the 'price_label' dataframe
        price_label = pd.concat([price_label, pd.DataFrame({'ID': [row['ID']], 'Average': [row['Average']], 'PRICE': [price_value], 'Price label': [label]})], ignore_index=True)

# Displaying the final dataframe 'price_label'
display(price_label)
```
